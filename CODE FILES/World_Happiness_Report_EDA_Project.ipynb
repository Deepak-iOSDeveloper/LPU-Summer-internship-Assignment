## 2. Standardize Column Names (hint)

```python
mapping = {
    'Country': 'Country',
    'Country or region': 'Country',

    'Happiness Score': 'Happiness',
    'Score': 'Happiness',
    'Happiness.Score': 'Happiness',

    'Happiness Rank': 'Happiness Rank',
    'Overall rank': 'Happiness Rank',
    'Happiness.Rank': 'Happiness Rank',

    'Economy (GDP per Capita)': 'GDP',
    'GDP per capita': 'GDP',
    'Economy..GDP.per.Capita.': 'GDP',

    'Family': 'Social Support',
    'Social support': 'Social Support',

    'Health (Life Expectancy)': 'Health',
    'Healthy life expectancy': 'Health',
    'Health..Life.Expectancy.': 'Health',

    'Freedom': 'Freedom',
    'Freedom to make life choices': 'Freedom',

    'Trust (Government Corruption)': 'Corruption',
    'Perceptions of corruption': 'Corruption',
    'Trust..Government.Corruption.': 'Corruption',

    'Generosity': 'Generosity',

    'year': 'year'
}

```

```python
dfs = [year_2015, year_2016, year_2017, year_2018, year_2019]
cleaned_dfs = []

for df in dfs:
    # 1. Keep only columns that exist in our mapping keys
    valid_cols = [col for col in df.columns if col in mapping]
    df_filtered = df[valid_cols].copy()

    # 2. Rename the remaining columns using your mapping
    df_filtered.rename(columns=mapping, inplace=True)

    # 3. Append to our clean list
    cleaned_dfs.append(df_filtered)

# Combine everything perfectly—no null values!
combined_df = pd.concat(cleaned_dfs, ignore_index=True)
```

### Conclusion
Countries like Mexico and Brazil showed an increase in social support while their happiness scores decreased. This indicates that while social support is crucial, it alone cannot guarantee rising happiness, especially if other factors like economic stability, freedom, or governance are deteriorating.
